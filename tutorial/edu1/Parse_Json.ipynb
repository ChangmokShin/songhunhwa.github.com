```python
import json
import pandas as pd
import numpy as np
```

```python
data= []
mydf = pd.DataFrame()

# to parse json files
with open('/Users/woowahan/Documents/Python/DS_Ext_School/edu1/ds_tutorial_json.json') as f:
    for line in f:
        data.append(json.loads(line))
        
    # convert json to df
    for i in range(0, len(data)):
        df = pd.DataFrame.from_dict([data[i]])
        mydf = mydf.append(df)
        
# reset index
mydf.reset_index(drop=True, inplace=True)
```
