```python
from pandasql import *
```

```python
q = """SELECT * 
       FROM df 
       WHERE ext = 'PDF'
           AND ismydoc = '0' 
       LIMIT 10"""
```

```python
print sqldf(q, locals())
```

```python
df[(df['ext'] == 'PDF') & (df['ismydoc'] == 0)].head(10)
```

```python
q1 = """SELECT
            ext,
            count(ext) as count,
            count(distinct sessionid) as unq_sess
        FROM df
        GROUP BY ext
        ORDER BY count DESC
     """
```

```python
print sqldf(q1, locals())
```

```python
df.groupby("ext").agg({"ext": "count", "sessionid": "nunique"}).sort_values("ext", ascending=False)
```
