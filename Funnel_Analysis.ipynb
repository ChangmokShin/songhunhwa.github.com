````python
# fill na using predictive model
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

ind_cols = ['actiontype', 'ismydoc', 'ext', 'screen']

# train without na
df_ms1 = df_ms.dropna()

# split ind, tar
df_ms_ind = df_ms1[ind_cols]
df_ms_tar = df_ms1[['documentposition']]
````

````python
df_ms_ind.head(10)
````

````python
df_ms_tar[:10]
````

````python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder
````

````python
df_ms_ind = df_ms_ind.apply(lambda x: encoder.fit_transform(x))

df_ms_ind.head(10)
````

````python
df_ms_tar = df_ms_tar.apply(lambda y: encoder.fit_transform(y))

df_ms_tar[:10]
````

````python
# for cross-val
# X: df_ms_ind / y: df_ms_tar

X_train, X_test, y_train, y_test = train_test_split(df_ms_ind, df_ms_tar, random_state=0)
````

````python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
````

````python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5).fit(X_train, y_train)

knn
````

````python
accuracy = knn.score(X_test, y_test)

print accuracy
````

````python
knn_pred = knn.predict(X_test)
knn_pred[:20]
````

````python
cm = confusion_matrix(y_test, knn_pred)
cm
````

### 신규  session_id 부여

````python
s = []
j = 0
for i in range(len(df)-1):
    if df.ix[i, 'sessionid'] == df.ix[i+1, 'sessionid']:
        s.append(j)
    else:
        s.append(j)
        j += 1
````

````python
df['new_session']  = pd.Series(['s' + str(x) for x in s])
````

````python
# fill missing values
df['new_session'] = df['new_session'].fillna(method='ffill')
````
